```python
import pandas as p
moviedata=p.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')
```

```python
import sqlite3
conn=sqlite3.connect('/kaggle/working/movieData.db')

moviedata.to_sql('moviedata',conn, if_exists='replace',index=False)
```

```python
%%sql
SELECT Director AS dir_name, Genre AS genre, SUM(Gross) AS gross
FROM moviedata
GROUP BY Director, Genre
ORDER BY gross DESC;
```

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('movieData.db')
Query=  """SELECT Director AS dir_name, Genre AS genre, SUM(Gross) AS gross
FROM moviedata
GROUP BY Director, Genre
ORDER BY gross DESC;"""
Temp_df= pd.read_sql(Query,conn)
Temp_df.to_csv("dir_Q1.csv", index= False)
```

```python
%%sql
select Genre as genre,Released_Year as R_year,count(*) as no_of_movies
from moviedata
group by Genre, Released_Year
order by Released_Year DESC, no_of_movies DESC;
```

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('movieData.db')
Query=  """select Genre as genre,Released_Year as R_year
FROM moviedata
group by Genre
order by Released_Year DESC;"""
Temp_df= pd.read_sql(Query,conn)
Temp_df.to_csv("genre_Q2.csv", index= False)
```

```python
%%sql
select IMDB_Rating as rating, Gross as gross,sum(IMDB_Rating) as s_rating,sum(Gross) as s_gross,
count(IMDB_Rating) as count_rating,count(Gross) as count_gross
from moviedata
group by IMDB_Rating,Gross
order by s_rating DESC,s_gross DESC,count_rating DESC,count_gross DESC;
```

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('movieData.db')
Query=  """select IMDB_Rating as rating, Gross as gross,sum(IMDB_Rating) as s_rating,sum(Gross) as s_gross,
count(IMDB_Rating) as count_rating,count(Gross) as count_gross
from moviedata
group by IMDB_Rating,Gross
order by s_rating DESC,s_gross DESC,count_rating DESC,count_gross DESC;"""
Temp_df= pd.read_sql(Query,conn)
Temp_df.to_csv("imbd_Q3.csv", index= False)
```

```python
%%sql
select Runtime as movie_len,IMDB_Rating as rating,avg(IMDB_Rating) as avg_rating,count(IMDB_Rating) as no_of_mov
from moviedata
where Runtime is not null and IMDB_Rating is not null
group by Runtime
order by avg_rating DESC;
```

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('movieData.db')
Query=  """select Runtime as movie_len,IMDB_Rating as rating,avg(IMDB_Rating) as avg_rating,count(IMDB_Rating) as no_of_mov
from moviedata
where Runtime is not null and IMDB_Rating is not null
group by Runtime
order by avg_rating DESC;"""
Temp_df= pd.read_sql(Query,conn)
Temp_df.to_csv("comp_Q4.csv", index= False)
```

```python
%%sql
select Runtime as movie_len,avg(Gross) as avg_gross,count(IMDB_Rating) as no_of_mov
from moviedata
where Runtime is not null and Gross is not null and IMDB_Rating is not null
group by Runtime
order by avg_gross DESC;
```

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('movieData.db')
Query=  """select Runtime as movie_len,avg(Gross) as avg_gross,count(IMDB_Rating) as no_of_mov
from moviedata
where Runtime is not null and Gross is not null and IMDB_Rating is not null 
group by Runtime
order by avg_gross DESC;"""
Temp_df= pd.read_sql(Query,conn)
Temp_df.to_csv("comp_Q4.1.csv", index= False)
```

```python
%%sql
select Star1,Gross as gross,IMDB_Rating as rating
from moviedata
where Gross is not null
group by Star1 
order by rating DESC;
```

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('movieData.db')
Query=  """select Star1,Gross as gross,IMDB_Rating as rating
from moviedata
where Gross is not null
group by Star1 
order by rating DESC;"""
Temp_df= pd.read_sql(Query,conn)
Temp_df.to_csv("actor_Q5.csv", index= False)
```
